15 like Science (S)

5 like both subjects

Ans. Mathematics or Science

n(M∪S)=20+15−5=30

n(M∪S)=20+15−5=30 30/50

 The probability that a student chosen at random likes Mathematics or Science is 3/5.

Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling? Why?



Stratified Sampling should be used because it ensures equal representation of both boys and girls by selecting samples separately from each group, whereas Simple Random Sampling may not provide balanced representation.

Question 10: Differentiate between discrete and continuous random variables with examples

Ans. Discrete random variables take countable values such as the number of students or defective items, whereas continuous random variables take uncountable values such as height, weight, or time which can be measured over a range.
